# Training process begins

finding best random state

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

# Best random state

```python
maxAccu=0 # maximum accuracy
maxRS =0 #best random state for which maximum accuracy is achieved

for i in range(1,100):
    pt_feature_train, pt_feature_test, target_train, target_test = train_test_split(pt_feature,target,test_size=20, random_state =i)
    LR = LogisticRegression()
    LR.fit(pt_feature_train, target_train)  # fitting the data trains the model
    predrf = LR.predict(pt_feature_test)  #this is the predicted target variable
    acc = accuracy_score(target_test, predrf)  # accuracy score
    print('accuracy_score', acc, 'random_state',i)
    
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
        print('accuracy', maxAccu, 'random_state',i)
```

```python
print('Best accuracy is',maxAccu, 'on Random state',maxRS)
```

# Ceating Train_test split

```python
pt_feature_train, pt_feature_test, target_train, target_test = train_test_split(pt_feature,target,test_size=.20, random_state= 17)
```

# Logistic Regression

```python
from sklearn.metrics import classification_report

LR = LogisticRegression()
LR.fit(pt_feature_train, target_train)
predlr = LR.predict(pt_feature_test)
print('Accuracy',accuracy_score(target_test,predlr)*100)
print(confusion_matrix(target_test,predlr))
print(classification_report(target_test,predlr))
```

# Support Vector Classifier

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(pt_feature_train, target_train)
pred = svc.predict(pt_feature_test)
print('Accuracy',accuracy_score(target_test,pred)*100)
print(confusion_matrix(target_test,pred))
print(classification_report(target_test,pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(pt_feature_train, target_train)
predrf = rf.predict(pt_feature_test)
print('Accuracy',accuracy_score(target_test,predrf)*100)
print(confusion_matrix(target_test,predrf))
print(classification_report(target_test,predrf))
```

# XG Boost

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report

XGB = XGBClassifier()
XGB.fit(pt_feature_train, target_train)
predxg = XGB.predict(pt_feature_test)
print('Accuracy',accuracy_score(target_test,predxg)*100)
print(confusion_matrix(target_test,predxg))
print(classification_report(target_test,predxg))
```

# K Nearest Neighbor  

```python
from sklearn.neighbors import KNeighborsClassifier

knn =  KNeighborsClassifier(7)
knn.fit(pt_feature_train, target_train)
predkn = knn.predict(pt_feature_test)
print('Accuracy',accuracy_score(target_test,predkn)*100)
print(confusion_matrix(target_test,predkn))
print(classification_report(target_test,predkn))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(pt_feature_train, target_train)
preddt = dt.predict(pt_feature_test)
print('Accuracy',accuracy_score(target_test,preddt)*100)
print(confusion_matrix(target_test,preddt))
print(classification_report(target_test,preddt))
```

# Cross Validation
