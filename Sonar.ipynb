# Training process begins

finding best random state

```python
# TRAINING ===> Bias created ===> selection bias
```

```python
#10 observation(5-Rock, 5-Mine) ===> 6 training(4 mine , 1 rock)(imbalance -- 4:1 or 80:20)
#4- testing( 5 - mine)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

# Best random state

```python
maxAccu=0 # maximum accuracy
maxRS =0 #best random state for which maximum accuracy is achieved

for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=20, random_state =i)
    LR = LogisticRegression()
    LR.fit(x_train, y_train)  # fitting the data trains the model
    predrf = LR.predict(x_test)  #this is the predicted target variable
    acc = accuracy_score(y_test, predrf)  # accuracy score
    print('accuracy_score', acc, 'random_state',i)
    
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
        print('accuracy', maxAccu, 'random_state',i)
```

```python
print('Best accuracy is',maxAccu, 'on Random state',maxRS)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

LR = LogisticRegression()
LR.fit(x_train, y_train)
predlr = LR.predict(x_test)
print('Accuracy',accuracy_score(y_test,predlr)*100)
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

#  Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
preddt = dt.predict(x_test)
print('Accuracy',accuracy_score(y_test,preddt)*100)
print(confusion_matrix(y_test,preddt))
print(classification_report(y_test,preddt))
```

# SVC

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train, y_train)
pred = svc.predict(x_test)
print('Accuracy',accuracy_score(y_test,pred)*100)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

# Random Forest Clssifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
predrf = rf.predict(x_test)
print('Accuracy',accuracy_score(y_test,predrf)*100)
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

logistic regression,Random forestand svc are producing good accuracy. Checking cross validation score for overfitting(if exist)

```python
from sklearn.model_selection import GridSearchCV

#creating parameter list
parameters ={'max_features':['auto','sqrt','log2'],
           'max_depth':[4,5,6,7,8],
           'criterion':['gini', 'entropy']}
```

```python
GCV= GridSearchCV(RandomForestClassifier(),parameters,cv=5,scoring='accuracy')
GCV.fit(x_train,y_train) #fitting the data 
GCV.best_params_  # Printing the best parameter found by GridSearchCV
```

```python
type(GCV)
```

```python
GCV.best_estimator_
```

```python
GCV_pred= GCV.best_estimator_.predict(x_test)  #predict with best parameter
accuracy_score(y_test,GCV_pred) # checking final accuracy
```

# ROC AUC plot
